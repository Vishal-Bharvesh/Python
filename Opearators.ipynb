Arithmatic : + - / * %

```python
digit=10;
print(digit)


digit += 5   #digit = digit + 5;
print(digit)

digit -= 10  # digit = digit - 10
print(digit)

digit *=10  # digit = digit * 10
print(digit)

digit /=5
print(digit)

print("-----------------")

digit=2;
print("DIGIT : ",digit)

digit **=digit;
print("**DIGIT : ",digit)


digit **=7;
print("**7 : ",digit)

digit=7*7*7*7*7*7*7
print("7*7*7*7*7*7*7 : ",digit)

```

Logical Operators : AND , OR , and NOT

AND  T and T = T
AND  T and F = F
AND  F and T = F
AND  F and F = F


---------------

OR  T and T = T
OR  T and F = T
OR  F and T = T
OR  F and F = F




```python
result = (6 > 3)
print(result)

result = (6 > 10)
print(result)



print("------AND-------")

result = (6 > 3) and  (6 > 10)
print(result)

print("------OR-------")


result = (6 > 3)  or (6 > 10)
print(result)

print("------NOT-------")


result =  not(  (6 > 3)  or (6 > 10)  )
print(result)

```
