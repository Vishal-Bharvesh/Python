```python
def getSquare(number):
  sq=number * number
  print("Arg Number : ",number)
  print("SQUARE : ",sq)
  return
```

```python
getSquare(5)
```

```python
getSquare(50)
```

```python
def getPowerResult(number,power):
  # 3 2 1 0
  actualpower=power
  result=1;
  while power>0:
    result *= number
    power -= 1;
  else:
     print("Arg Number : ",number)
     print("Arg Power : ",actualpower)
     print("RESULT : ",result)
  return

```

```python
number = int(input("Enter Digit : "))
power = int(input("Enter Power Value : "))

getPowerResult(number,power)
```
