LIST DataType

------------
List req : "[ ]"

```python
mylist1=["Python","Solidity","React.js","Node.js"]
```

```python
print(mylist1)
```

```python
print(type(mylist1))
```

```python
print(mylist1[0])
```

```python
cnt=0
#totallistitems=mylist1.len()
totallistitems=len(mylist1)

while cnt < totallistitems:
  print(cnt," : ",mylist1[cnt])
  cnt += 1
```

```python
mylist2=[1,2,3,4,5,6]
print(mylist2)
print(type(mylist2))
```

```python
mydetails=["Smith",45,'Male',165.4,78.5]
print(mydetails)
```

```python
cnt=0
#totallistitems=mylist1.len()
totallistitems=len(mydetails)

while cnt < totallistitems:
  mytype=type(mydetails[cnt])
  print(cnt," : ",mydetails[cnt]," - ",mytype)
  cnt += 1
```
