For : Seq.

```python
print("BOP")
print("--------------")

vstring=input("Entry Some Text : ")

for vchar in vstring:   # foreach {char} in {string}
  print(vchar)


print("--------------")
print("EOP")
```

BREAK

```python
print("BOP")
print("--------------")

number=1;

while True:
  print("Number : ",number)
  if number == 11:
    break # to terminate the Parent Loop | break steme comes under Jumping Stmt.
  number += 1

print("--------------")
print("EOP")
```
