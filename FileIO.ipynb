```python
myfile = open('/drive/My Drive/Colab Notebooks/myaddress02.pdf','r') # r = read
filedata=myfile.readlines() # LISTS
print(filedata)
myfile.close();
```

```python
myfile = open('/drive/My Drive/Colab Notebooks/myaddress02.pdf','r') # r = read
filedata=myfile.readlines() # LISTS

totallines = len(filedata)
line=0

while line < totallines:
  print(filedata[line])
  line += 1


myfile.close();
```

```python
# File Reading with FOR LOOP

myfile = open('/drive/My Drive/Colab Notebooks/myaddress02.pdf','r') # r = read
for line in myfile:
  print(line)

myfile.close();
```
