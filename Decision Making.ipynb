```python
digit= int( input("Enter a Digit : ") )
print("Value : ",digit)
print("-------")

# IF

if digit==1:
  print("Yes, Value is ONE")

print("EOP")
```

```python
digit= int( input("Enter a Digit : ") )
print("Value : ",digit)
print("-------")

# IF...ELSE

if digit==1:
  print("Yes, Value is ONE")
else:
  print("NO, Value is Not ONE")

print("EOP")
```

```python
digit1= int( input("Enter First Digit : ") )
digit2= int( input("Enter Second Digit : ") )


# IF..elif...

if digit1 > digit2:
  print(digit1," is MAX")   #TRUE STMT
elif digit2 > digit1:
  print(digit2," is MAX") # FALSE STMT
else:
  print("Both Values are SAME");

print("EOP")
```

```python
print("BOP")
print("--------------")
digit1= int( input("Enter First Digit : ") )
digit2= int( input("Enter Second Digit : ") )
digit3= int( input("Enter Third Digit : ") )

if (digit1 > digit2) and (digit1 > digit3):
  print(digit1," is MAX")
elif digit2 > digit3:    #(digit2 > digit1) and (digit2 > digit3):
  print(digit2," is MAX")
elif digit3 > digit2:
  print(digit3, "is MAX")
else:
  print("ALL Values as SAME")


print("--------------")
print("EOP")
```
